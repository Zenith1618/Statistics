## T-Test

A t-test is a type of inferential statistic which is used to determine if there is a significant difference between the means of two groups which may be related in certain features

T-test has 2 types : 1. one sampled t-test 2. two-sampled t-test.

####  One-sample T-test with Python

The test will tell us whether means of the sample and the population are different

image.png

```python
import numpy as np
```

```python
ages=[10,20,35,50,28,40,55,18,16,55,30,25,43,18,30,28,14,24,16,17,32,35,26,27,65,18,43,23,21,20,19,70]      #Population
```

```python
ages_mean = np.mean(ages)
print(ages_mean)
```

```python
# Lets contruct a sample and test out whether the sample does come from population
sample_size = 10
age_sample = np.random.choice(ages, sample_size)
```

```python
age_sample
```

```python
age_sample_mean = np.mean(age_sample)
print(age_sample_mean)
```

It is close, but lets test it out

```python
from scipy.stats import ttest_1samp     #T-test for 1 sample
```

```python
ttest_1samp(age_sample, popmean = 30.34, alternative='two-sided')
#‘two-sided’: the mean of the underlying distribution of the sample is different than the given population mean (popmean)
```

As you can see pValue > 0.05(alpha) we can say that, we accept the Null Hypothesis, which indicates that, the sample does come from population

#### **Given class mean age of the student(sample), does it comes from school mean age of students(population)**

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
```

```python
np.random.seed(6)
school_ages = stats.poisson.rvs(loc=18, mu=35, size=1500)   #start from 18, where mean = 35 and total points = 1500
class_ages = stats.poisson.rvs(loc=18, mu=30, size=60)
```

```python
school_ages
```

```python
class_ages
```

```python
class_ages.mean()
```

```python
ttest_1samp(class_ages, popmean = school_ages.mean(), alternative='two-sided')
```

#### As pValue > .05(alpha) we accept the Null Hypothesis, that is the sample comes from population
